## 1.10.リスト

本編: [1.10.リスト](https://atcoder.jp/contests/APG4bPython/tasks/APG4bPython_k)

練習問題: [EX10.平均との差](https://atcoder.jp/contests/APG4bPython/tasks/APG4bPython_cm)

これまでは変数1つにつき値を1つしか入れられませんでした。
リスト (配列) を使うと、複数の値をまとめて1つの変数で管理することができます。
for文と組み合わせるとめちゃくちゃ強力なので、ここで使い方をしっかり覚えましょう。

```python
a = [3, 1, 4, 1, 5]  # リストの作成
print(a)
print(len(a))  # リストの要素数

print(a[0])   # 最初の要素 (0番目)
print(a[2])   # 3番目の要素 (2番目)
print(a[-1])  # 最後の要素

print(a[1:3])  # 1番目から2番目まで (3番目は含まない) のスライス
print(a[:2])   # 先頭から1番目まで
print(a[2:])   # 2番目から最後まで
```

### 注意点

- リストの添字 (インデックス) は **0から始まる** ことに注意しましょう
- 存在しない要素を指定すると `IndexError` というエラー (AtCoderでは RE: Runtime Error) になります
- `a[-1]` のように負の添字を使うと、後ろから数えた要素を取得できます

### for文との組み合わせ

`range()` はリストに変換することもできます。
リストは for文で1つずつ取り出して処理するのが一番よく使う使い方です。

```python
b = list(range(5))  # [0, 1, 2, 3, 4]
print(b)

for x in b:
    print(x)
```

### リストへの入力

`map` 関数を使うと、`split()` で分割した文字列をまとめて `int` に変換できます。
これでN個の数値をまとめてリストとして受け取ることができます。

```python
# "1 2 3 4 5" のような入力を、空白区切りの整数のリストとして受け取る
a = list(map(int, input().split()))
print(a)

# 合計と平均を求める
total = sum(a)
print(total)
print(total // len(a))  # 平均 (小数点以下は切り捨て)
```

### リストの操作

リストには便利な操作 (メソッド) がたくさん用意されています。
代表的なものをいくつか見てみましょう。

```python
a = [3, 1, 4, 1, 5]

a.append(9)  # 末尾に要素を追加
print(a)

a.pop()  # 末尾の要素を削除
print(a)

a.sort()  # 昇順に並び替え
print(a)

a.reverse()  # 順番を反転
print(a)

print(1 in a)      # 1 が a に含まれているか
print(a.count(1))  # 1 が a に何個含まれているか
```

### 発展: リストの代入は「コピー」ではない

Chapter 1 で学んだ通り、リストは可変オブジェクトです。
`b = a` とすると `a` と `b` は同じリストを指すため、片方を変更すると両方が変わってしまいます。
本当にコピーしたい場合は `b = a.copy()` を使いましょう。

```python
a = [1, 2, 3]
b = a
b.append(4)
print(a, b)  # b だけでなく a も変わってしまう

c = a.copy()
c.append(100)
print(a, c)  # c だけが変わる
```

それでは練習問題を解いてみましょう。
練習問題: [EX10.平均との差](https://atcoder.jp/contests/APG4bPython/tasks/APG4bPython_cm)

ヒント: 平均は `合計 // 人数` で求められます (小数点以下切り捨て)。
各人の点数と平均との「差」を求めるには `abs()` 関数 (絶対値) が便利です。

## 1.11.文字列

本編: [1.11.文字列](https://atcoder.jp/contests/APG4bPython/tasks/APG4bPython_l)

練習問題: [EX11.足したり引いたり](https://atcoder.jp/contests/APG4bPython/tasks/APG4bPython_cl)

文字列 (str型) にもリストと同じように、添字でアクセスしたりスライスしたりすることができます。
さらに文字列専用の便利なメソッドもたくさんあるので、ここで一気に紹介します。

s = "Hello, World"
print(len(s))  # 文字列の長さ
print(s[0])    # 先頭の文字
print(s[-1])   # 最後の文字
print(s[0:5])  # スライス
print(s + "!")  # 文字列の連結

### エスケープシーケンス

`\` を使うと、`"` 自身や改行など特殊な文字を文字列の中に入れることができます。

```python
print("KCLC\n部活動")  # \n で改行
print("\"KCLC\"")        # \" で " を文字として表示
print("KCLC\t部活動")  # \t でタブ
```

### 検索・判定系のメソッド

文字列の中に特定の文字が含まれているか、何文字目にあるかなどを調べるメソッドです。

```python
s = "Hello, World"

print("World" in s)      # 含まれているか
print(s.index("World"))  # 何文字目から含まれているか (0始まり)
print(s.count("l"))       # 出現回数

print(s.isupper())      # 全部大文字か
print(s.islower())      # 全部小文字か
print("123".isdigit())  # 全部数字か
```

### 変換系のメソッド

大文字・小文字の変換や、分割・結合・置換もよく使います。

```python
s = "Hello, World"
print(s.upper())  # 大文字に変換
print(s.lower())  # 小文字に変換

words = s.split(", ")  # 文字列を分割してリストに
print(words)

print("-".join(words))  # リストを文字列に結合

print(s.replace("World", "KCLC"))  # 置換
```

### 数値⇔文字列の変換、Unicode

`int()` / `str()` で数値と文字列を変換できることは Chapter 1 でも触れましたね。
`ord()` / `chr()` を使うと、文字と文字コード (Unicode のコードポイント) を変換できます。

```python
n = int("123")  # 文字列 -> 数値
print(n + 1)

s = str(123)  # 数値 -> 文字列
print(s + "円")

print(ord("A"))  # 文字 -> 文字コード
print(chr(65))   # 文字コード -> 文字
```

### 文字列は書き換えられない (immutable)

文字列は不変オブジェクトのため、`s[0] = "M"` のように一部分だけ書き換えることはできず、エラーになります。
1文字だけ変えたい場合は、いったん `list()` でリストに変換してから書き換え、`join` で文字列に戻しましょう。

それでは練習問題を解いてみましょう。
練習問題: [EX11.足したり引いたり](https://atcoder.jp/contests/APG4bPython/tasks/APG4bPython_cl)

ヒント: 1文字ずつfor文で見ていき、`+` なら結果に1を足し、`-` なら結果から1を引きましょう。
入力は最初と最後が必ず `1` なので、最初の `1` を初期値として、2文字目以降を2文字ずつ (符号, 数字) 見ていくと書きやすいです。
