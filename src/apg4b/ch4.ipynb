# 1.12 組み込み関数
そもそも関数とは何でしょうか。
数学では$x$を決めると$y$が一つに決まるとき$y$は$x$の関数であると言ったりします。
もちろんプログラミングにおける関数でもそのようなものはあります。
しかしプログラミングではより広く、ある処理の枠組みを広く関数と呼んでいます。

1.13で扱う関数は自分で実装するものです。
しかしこの章で扱う組み込み関数とはPythonの開発者が用意したもので、認識としては
便利な機能程度にとらえておくとよいです。
では、いくつか例を見ていきましょう。

| 関数 | 機能 |
| --- | --- |
| ```abs(a)``` | aの絶対値を返す |
| ```pow(a,b) ``` | aのb乗を返す |
| ```pow(a,b,mod)``` | aのb乗をmodで割った余りを返す |
| ```max(a,b)``` | aとbのうち小さくないほうを返す |
| ```min(a,b)``` | aとbのうち大きくないほうを返す |
| ```sum(a)``` | 配列aの要素の総和を返す|
| ```sorted(a)``` | 配列aをソートした(大小に沿って並べ替えた)物を返す|
| ```all(条件式1,条件式2,...,)```| 条件式1,2,...がすべて真であるかを返す|
| ```any(条件式1,条件式2,...,)```| 条件式1,2,...の中に真であるものが一つ以上あるかを返す|
| ```enumrate(a)``` | ```for```文の中で配列aの添え字と要素を同時に取得 |

# 1.13 関数
では、関数を自作してみましょう。
関数を実装するときは以下のように書きます
```python
def [関数名]([引数]):
  [処理]
  [処理]
  ...
  return [返り値]

def [別の関数]
  [処理]
  ...

[メインの処理が始まる]
```
覚えておいてほしいのは
- 関数を定義するときは```def```を最初に書いて今から関数を用意することを宣言
- 関数名の右に処理のために与えるパラメータ(**引数**と呼ばれる)をカッコ内に列挙する
- 関数内の処理は1回インデントしてから行ごとに書いていく
- 最後に```return```と書き、その右に返すもの(**返り値**)を書いて値を返すことを宣言

という点です。
引数や返り値はなくてもいいです。
しかしfor文などと同様にインデントは絶対に忘れないでください。

正直、初めのころはあまり関数を使うメリットは享受できないかもしれませんが
今後、ゲーム開発などで多くの量の処理を実装するときにコードの構造が明確になりやすい(どの処理をかたまりとして行っているかが明快になる)ので覚えておきましょう。
