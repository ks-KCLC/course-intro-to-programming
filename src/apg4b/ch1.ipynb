### だいぶ発展: 変数はオブジェクトに紐づいている。

#### オブジェクトの種類

- 不変オブジェクト (ex. `int`, `float`, `str`) ※ インターン化 (興味あったら調べてみて)
- 可変オブジェクト (ex. `list`, `dict`, `set`)

#### オブジェクトの確認の仕方

- `id(x)` : 紐づいているオブジェクトのアドレスを確認
- `==` 演算子 : 値が等しいか
- `is` 演算子 : 同じオブジェクトを参照しているか (`id(a))

```python
# 不変オブジェクト vs 可変オブジェクト
a = 1
b = a
a += 1
print(a, b)  # a と b は異なる値を持つ (不変オブジェクト、再代入されると新しいオブジェクトが作られる)
a = [1, 2, 3]
b = a
a.append(4)
print(a, b)  # a と b は同じ値を持つ (可変オブジェクト、再代入されると同じオブジェクトが更新される)
```

```python
# インターン化
a = "AI"
b = "AI"
print(a is b)  # a と b は同じオブジェクトを参照している (インターン化される文字列、違ったらごめん)
a = "Hello, World"
b = "Hello, World"
print(a is b)  # a と b は異なるオブジェクトを参照している (インターン化されない文字列、違ったらごめん)
```
