## 2.01 いろいろなfor文

本編: [2.01 いろいろなfor文](https://atcoder.jp/contests/APG4bPython/tasks/APG4bPython_p)

第一章ではfor文を学習しましたが、それをさらに発展させたりした色々な書き方を紹介します。

### rangeを使う書き方

1.09で扱った内容の復習です。```range関数```を使うことで指定した範囲のループ処理ができます。

#### range(stop)

````python
# 0 以上 5 未満をループ
for i in range(5):
    print(i)
````

#### range(start,stop)

例えば、 2以上 5未満の範囲をループしたい時は、()の中を```(最初の数、終わりの数)```とすることで実行する事ができます。

````python
# 2 以上 5 未満をループ
for i in range(2, 5):
    print(i)
````

#### range(start, stop, step)

2以上10未満の範囲を2ずつ増やしながらループ、といった処理もする事ができます。
()の中に```(最初の数,終わりの数,増やす数)```と入力することで実行する事ができます。

````python
# 2 以上 10 未満の範囲を 2 ずつ増やしながらループ
for i in range(2, 10, 2):
    print(i)
````

###  リストなどに対するfor文

#### リストに対するfor文
リストに対するfor文は1.10でも扱った通りです。

````python
L = [10, 20, 40, 80, 160]

# L の要素を順に出力
for x in L:
    print(x)
````

#### 文字列に対するfor文
文字列に対してfor文を使う事で、1文字ごとに処理する事ができます。

````python
S = "AtCoder"

# s の要素を順に出力
for x in S:
    print(x)
````

**実行結果**

````python
A
t
C
o
d
e
r
````

### リストのリスト

#### リストのリスト
リストを要素とするリストの場合などには、カンマ区切りで受け取ると複数の変数に一気に格納することができます。

````python
L = [[5, 10], [6, 20], [7, 50]]
for a, b in L:
    print("a は", a, "です。 b は", b, "です。")
````

**実行結果**

````python
a は 5 です。 b は 10 です。
a は 6 です。 b は 20 です。
a は 7 です。 b は 50 です。
````

#### 3 要素以上の場合
内側のリストが 3 要素以上になっても同様に書けますが、受け取る側の引数の個数が異なるとエラーになるので注意してください。

````python
L = [[1, 2, 3], [4, 5, 6]]
for a, b, c in L:
    print("a は", a, "です。 b は", b, "です。 c は", c, "です。")
````

**実行結果**

````python
a は 1 です。 b は 2 です。 c は 3 です。
a は 4 です。 b は 5 です。 c は 6 です。
````

次の例では長さ 4 の配列を 3 つの変数```(a, b, c)```で受け取ろうとしているため、 ```ValueError: too many values to unpack (expected 3)``` というエラーになります。

````python
L = [[1, 2, 3, 4], [5, 6, 7, 8]]
for a, b, c in L:
    print("a は", a, "です。 b は", b, "です。 c は", c, "です。")
````

#### enumerate
enumerate 関数を使うと、リストのインデックス(何番目の数か)と要素をまとめて取得できます。

````python
L = [10, 20, 40, 80, 160]

# L のインデックスと要素を取得
for i, x in enumerate(L):
    print(i, "番目の要素は", x)
````

**実行結果**

````python
0 番目の要素は 10
1 番目の要素は 20
2 番目の要素は 40
3 番目の要素は 80
4 番目の要素は 160
````

この場合、インデックスは0から始まり0,1,2,3...と続いていきます。最初のインデックスを指定したい場合、```enumerate(L,最初の数字)```とします。

````python
L = [10, 20, 40, 80, 160]

# インデックスを 10 から開始
for i, x in enumerate(L, 10):
    print(i, "番目の要素は", x)
````

**実行結果**

````python
10 番目の要素は 10
11 番目の要素は 20
12 番目の要素は 40
13 番目の要素は 80
14 番目の要素は 160
````

#### リストを逆順に処理
```reversed```関数を使うと、リストを逆順に処理できます。
0番目,1番目,2番目と処理していたものが、2番目,1番目,0番目と処理する事ができます。

````python
L = [10, 20, 40, 80, 160]

# L の逆順
for a in reversed(L):
    print(a)
````

**実行結果**

````python
160
80
40
20
10
````

#### zip
複数のリストを見てまとめてループできるのがzip関数です。
配列Aと配列Bがある時zip関数を使うことで、Aの1番目とBの1番目、Aの2番目とBの2番目...というように処理してくれます。

````python
A = [3, 4, 5, 6, 7]
B = [10, 20, 40, 80, 160]

# A と B から 1 つずつ要素を取得し a および b に格納する
for a, b in zip(A, B):
    print(a, b)
````

**実行結果**

````python
3 10
4 20
5 40
6 80
7 160
````

リストのサイズが異なる場合、短い方が終わるまで処理されます。

````python
A = [3, 4, 5, 6, 7, 8, 9, 10]
B = [10, 20, 40, 80, 160]

# A と B から 1 つずつ要素を取得し a および b に格納する
for a, b in zip(A, B):
    print(a, b)
````

````python
3 10
4 20
5 40
6 80
7 160
````

### continue,break,else
1.09でやったことの復習です。この機に思い出しましょう。

#### continue
continueはその後の処理を飛ばしてループの先頭に戻ります。

````python
for i in range(5):
    # i == 2 の場合はその後の処理を飛ばす
    if i == 2:
        continue
    print(i)
````

````python
0
1
3
4
````

#### break
breakを使うとループを抜けてもうそのループの処理は行われなくなります。

````python
for i in range(5):
    # i == 2 の場合はループを抜ける
    if i == 2:
        break
    print(i)
````

````python
0
1
````

#### break-else
breakを含む文の直後にelseをおくと、breakが発動しなかった時にelseが処理されます。
例えば、次のコードはbreakが発動したのでelseは処理されません。

````python
for i in range(5):
    if i == 2:
        break
    print(i)
else:
    print("else 節が処理されたよ")
````

````python
0
1
````

次の文はbreakで抜けないので、elseが処理されます。

for i in range(5):
    if i == 7:
        break
    print(i)
else:
    print("else 節が処理されたよ")


````python
0
1
2
3
4
else 節が処理されたよ
````

## 2.02.多重ループ
ここからはプログラミングでよく使う多重ループについて学びます。
実務や競プロなど様々な分野で確実に使うものなのでしっかり身につけましょう。

#### 多重ループ

1.08や1.09で扱ったwhile文、for文などのループは、さらにその中にループを書くことができます。

````python
for i in range(3):
    for j in range(3):
        print(i, j)
````

このように、ループの内側にループがあるようなもののことを二重ループと呼びます。
二重だけではなく、三重、四重と重ねることもできます。これらをまとめて多重ループと言います。

````python
for i in range(2):
    for j in range(2):
        for k in range(2):
            print(i, j, k)
````

#### 多重ループのbreak/continue

for文では先ほどやった通り、breakとcontinueという命令がありました。多重ループでも当然使うことはできます。
しかし、多重ループ内でbreakやcontinueを使ったとしても、1段階しかループを抜けることはできません。あくまでbreakを処理した内側のループ文から抜けるだけだからです。
例えば下のコードのようになってしまいます。


````python
for i in range(3):
    for j in range(3):
        print(i, j)
        if i == 1 and j == 1:
            print("i と j が両方 1 なので終了（？）")
            break
````

````python
0 0
0 1
0 2
1 0
1 1
i と j が両方 1 なので終了（？）
2 0
2 1
2 2
````

　多重ループをまとめて抜ける方法として最も手っ取り早いのは、「フラグ変数」と呼ばれるループを抜けるかどうかを管理するboolの変数を用意し、フラグ変数がtrueの時にループを抜けると言う方法です。

````python
# 外側のループを抜ける条件を満たしているかどうか（フラグ変数）
finished = False

for i in range(3):
    for j in range(3):
        print(i, j)
        if i == 1 and j == 1:
            print("i と j が両方 1 なので終了。")
            finished = True

        if finished:
            break

    if finished:
        break
````

````python
0 0
0 1
0 2
1 0
1 1
i と j が両方 1 なので終了。
````
