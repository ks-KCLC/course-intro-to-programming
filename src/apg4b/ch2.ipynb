## 1.06.if文・比較演算子・論理演算子

本編: [1.06.if文・比較演算子・論理演算子](https://atcoder.jp/contests/APG4bPython/tasks/APG4bPython_g)

ifを用いると指定した条件に当てはまるときだけ処理をするプログラムを書くことができます。
また、「比較演算子」というものについてはatcoderサイトを見ながらプログラムを作ってみましょう
下はif文の使い方の例になります、とてつもなく大事な文法なので必ず理解しておきましょう

```python
x=int(input())
if x<100:
    print("xは100より小さい")
elif x==100:
    print("xは100である")
else:
    print("xは100より大きい")
```

### 注意点と比較演算子
ifの後の:を忘れないようにしましょう
elifとelseは使いこなせるようにしましょう、超便利です
あとインデントを開けることを気をつけましょう
次に比較演算子について学習しましょう
計算の演算子とは間違えないように注意してください
"=="と"!="以外は直感的にわかると思うのでこの2つだけ覚えるようにしましょう
a!=bでaがbと異なること、a==bでaとbが等しいこと

```python
a,b=input().split()
a=int(a)
b=int(b)
if a==b:
    print("aとbは等しい")
elif a!=b:
    print("aとbは異なる")
else:
```

### 比較演算子＋計算と比較演算子の連鎖
比較演算子と計算式は両立することができます、次の例を見てください

```python
a,b=input().split()
c=int(input())
a=int(a)
b=int(b)
if a+b<c:
    print("aとbの和はcより小さい")
elif a+b==c:
    print("aとbの和はcである")
else:
    print("aとbの和はcより大きい")
```

### 論理演算子の活用
これで計算式を含めた条件式が書けるようになりましたね。次に「かつ」や「または」といった条件をつけたくなりますよね、これらを付属するのが「論理演算子」になります、超便利なので覚えましょう
aかつbがa and b
aまたはbがa or bです。
下のコードがandの活用方法です、orも同様に活用します

```python
a,b=input().split()
a=int(a)
b=int(b)
if a+b!=0 and 0<a+b:
    print("aとbの和は正")
elif a+b!=0 and 0>a+b:
    print("aとbの和は負")
else:
    print("aとbの和は0")
```

それでは練習問題を解いてみましょう、今回は割と難しめです

## 1.07.論理式の値とbool型

本編: [1.07.論理式の値とbool型](https://atcoder.jp/contests/APG4bPython/tasks/APG4bPython_h)

練習問題: [bool値パズル](https://atcoder.jp/contests/APG4bPython/tasks/APG4bPython_cp)

実は今までやってきた条件式ですが、式自体はbool型というので判別されています
たとえば
if a+b==5:という条件式なら、a+bが５の時のみbool型でのtrue(つまり真)となりそれ以外の時はfalse(偽)ということになります、つまりif文は「指定された式がtrueのときのみ」という条件を持っているということです。
実際に条件式がbool型をもっていることを確認してみましょう

```python
x = 0
print(x == 0) # 「x が 0 と等しいかどうか」を出力
print(x >= 1) # 「x が 1 以上かどうか」を出力
```

次に前述したifの仕組みを見ていきましょう

```python
if True:
    print("条件が成り立っています")
if False:
    print("条件が成り立っています")
else:
    print("条件が成り立っていません")
```

ちなみに数値の場合、0のみtrue、文字列の場合は""(空文字列)のみがfalseとなりますが、数値のみでの条件分岐はあまり使わないと思います（大体if文の中では条件式を用いると思うので）
それでは練習問題を解いてみましょう

## 1.09.for文・break・continue 

本編: [1.09.for文・break・continue ](https://atcoder.jp/contests/APG4bPython/tasks/APG4bPython_j)

練習問題: [電卓を作ろう２](https://atcoder.jp/contests/APG4bPython/tasks/APG4bPython_cn)

いよいよお待ちかねのfor文です、自分的にはwhileより絶対的に扱いやすいループの書き方です
これは本当にこれからすべてのところで使うので絶対に習得しましょう！


```python
# 2から10までの整数を一つずつ出力するコード
for i in range(2,11):
    print(i)
# 0から10までの整数を一つずつ出力するコード、range(0,11)としなければいけないが0の時のみ省略可
for i in range(10):
    print(i)
```

for文は理解できましたでしょうか？
本当に大事な文法なのでここで絶対に理解しましょう、わからなかったら先輩に気兼ねなく聞いてください
### breakとcontinue
breakはfor文を途中で抜けることが、conitnueは後の処理を飛ばして次のループへ進むことができます

```python
# break がなければこのループは i == 4 まで繰り返す
for i in range(5):
    if i == 3:
        print("ぬける")
        break  # i == 3 の時点でループから抜ける
 
    print(i)
 
print("終了")
```

### 2つ飛ばしのループなど
実は先ほど使ったrange(0,10)のような関数は0以上10未満の数列を生成するものです
forの正体は数列の中を最初から探索するものだったんですよね、これを利用して、iを２ずつ増やしてfor文を書くこともできます
range(a,b,c)でa以上b未満でcずつ増えていく数列を生成することができます。

```python
for i in range(0, 10, 2):
    print(i)
```
